```python
pd.options.mode.chained_assignment = None  # default='warn'

pd.set_option('display.max_columns', None)

df_box = pd.read_sql_query("SELECT * FROM box_scores", conn)

# Create new columns in box_scores 
df_box['Games'] = 1
df_box['Win Total'] = 0
df_box['Loss Total'] = 0
df_box['Score Against'] = 0
df_box['Estimated Possessions'] = 0
df_box['Estimated Possessions Against'] = 0
df_box['Points Per 100 Possessions'] = 0
df_box['Points Per 100 Possessions Allowed'] = 0
df_box['O Rebounds Allowed'] = 0
df_box['D Rebounds Allowed'] = 0

df_box['FGM Against'] = 0
df_box['3FGM Against'] = 0
df_box['FGA Against'] = 0
df_box['TOV Against'] = 0
df_box['FTA Against'] = 0
df_box['FTM Against'] = 0





# Creating an estimation for the amount of possessions a team has in a game   
df_box['Estimated Possessions'] = (df_box['FGA'] - df_box['ORB'] + .475*df_box['FTA']).astype(int)
   
                                                       

for i in range(len(df_box)-1):
    if i%2 == 0:   
        # Track defensive and metric stats based on team rates and opponent values in every game 
        if df_box['Score'][i]>df_box['Score'][i+1]:
            df_box['Win Total'][i] = 1
            df_box['Win Total'][i+1] = 0
            df_box['Loss Total'][i] = 0
            df_box['Loss Total'][i+1] = 1
            
        else:
            df_box['Win Total'][i] = 0
            df_box['Win Total'][i+1] = 1
            df_box['Loss Total'][i] = 1
            df_box['Loss Total'][i+1] = 0
            
        
        df_box['Score Against'][i] = df_box['Score'][i+1]
        df_box['Score Against'][i+1] = df_box['Score'][i]
        df_box['Estimated Possessions Against'][i] = df_box['Estimated Possessions'][i+1]
        df_box['Estimated Possessions Against'][i+1] = df_box['Estimated Possessions'][i]
        df_box['O Rebounds Allowed'][i] = df_box['ORB'][i+1]
        df_box['O Rebounds Allowed'][i+1] = df_box['ORB'][i]
        df_box['D Rebounds Allowed'][i] = df_box['DRB'][i+1]
        df_box['D Rebounds Allowed'][i+1] = df_box['DRB'][i]
        df_box['FGM Against'][i] = df_box['FGM'][i+1]
        df_box['FGM Against'][i+1] = df_box['FGM'][i]
        df_box['3FGM Against'][i] = df_box['3FGM'][i+1]
        df_box['3FGM Against'][i+1] = df_box['3FGM'][i]
        df_box['FGA Against'][i] = df_box['FGA'][i+1]
        df_box['FGA Against'][i+1] = df_box['FGA'][i]
        df_box['TOV Against'][i] = df_box['TOV'][i+1]
        df_box['TOV Against'][i+1] = df_box['TOV'][i]
        df_box['FTA Against'][i] = df_box['FTA'][i+1]
        df_box['FTA Against'][i+1] = df_box['FTA'][i]
        df_box['FTM Against'][i] = df_box['FTM'][i+1]
        df_box['FTM Against'][i+1] = df_box['FTM'][i]
                                           
        
        
        
df_box['Points Per 100 Possessions'] = (df_box['Score']/df_box['Estimated Possessions']*100).astype(int)
df_box['Points Per 100 Possessions Allowed'] = (df_box['Score Against']/df_box['Estimated Possessions Against']*100).astype(int)


# move 'Score Against' stat next to 'Score'
mid = df_box['Score Against']
df_box.drop(labels=['Score Against'], axis=1, inplace = True)
df_box.insert(4, 'Score Against', mid)
del df_box['GameId']
del df_box['Home']



df_box
```

```python
df_box = df_box.rename(columns={'Score': 'Points'})

df_box = df_box.rename(columns={'Score Against': 'Points Against'})

# sum up counting stats for each team's for and against values
df = df_box.groupby(['Team']).sum()

# showing points per 100 possessions (and against) average per game
df['Points Per 100 Possessions'] = (df['Points Per 100 Possessions'] / df['Games']).astype(int)

df['Points Per 100 Possessions Allowed'] = (df['Points Per 100 Possessions Allowed'] / df['Games']).astype(int)


# creating advanced stats based on various offensive and defensive rates

df['eFG Rate'] = ''
df['Turnover Rate'] = ''
df['FT Rate'] = ''
df['O Rebound Rate'] = ''
df['D Rebound Rate'] = ''

df['eFG Rate Against'] = ''
df['Turnover Rate Against'] = ''
df['FT Rate Against'] = ''
df['O Rebound Rate Against'] = ''
df['D Rebound Rate Against'] = ''

 

df['eFG Rate'] = (df['FGM'] + .5 * df['3FGM']) / (df['FGA'])
df['Turnover Rate'] = (df['TOV'])/(df['FGA']+.45*df['FTA']+df['TOV'])
df['FT Rate'] = df['FTM']/df['FGA']
df['O Rebound Rate'] = df['ORB']/(df['ORB'] + df['D Rebounds Allowed'])
df['D Rebound Rate'] = df['DRB']/(df['DRB']+df['O Rebounds Allowed'])

df['eFG Rate Against'] = (df['FGM Against'] + .5 * df['3FGM Against']) / (df['FGA Against'])
df['Turnover Rate Against'] = (df['TOV Against'])/(df['FGA Against']+.45*df['FTA Against']+df['TOV Against'])
df['FT Rate Against'] = df['FTM Against']/df['FGA Against']
df['O Rebound Rate Against'] = df['O Rebounds Allowed']/(df['O Rebounds Allowed'] + df['DRB'])
df['D Rebound Rate Against'] = df['D Rebounds Allowed']/(df['D Rebounds Allowed']+df['ORB'])


df



```

```python
# normalizing every stat to be on a scale where 100 is the stat's mean (eg. if team A has 100 points they would be at a conference average level for points)
# Similar to ERA+, each stat takes on a conference average of 100 (over 100 is above average, under 100 is below average)
# makes combining them later into one stat simpler because they are now all on the same levels
# I can now decide more easily how to weight each of them 

df_means = df.mean()

df = df * (100 / df_means)

df
```

```python
df['Rating'] = ''

# All stats in formula have already been normalized to be on same scale as each other (mean of 100)
# Creating a rating to rank teams based on predictive success
# Rating is created as follows:
# 25% net points per 100 possessions
# 20% net eFG rate
# 20% net FT rate
# 20% net turnover rate
# 10% net D rebound rate
# 5% net O rebound rate



df['Rating'] = .25*((df['Points Per 100 Possessions']-df['Points Per 100 Possessions Allowed'])) + .2*(df['eFG Rate']-df['eFG Rate Against'])   + .2*(df['FT Rate']-df['FT Rate Against']) + .2*(df['Turnover Rate Against']- df['Turnover Rate'])   + .1*(df['D Rebound Rate']-df['D Rebound Rate Against']) + .05*(df['O Rebound Rate']-df['O Rebound Rate Against']) 


# normalize each team on a scale of 0 to 1 to show how close each team is to the poles of the 'isolated' conference
df['Rating'] = (df['Rating'] - min(df['Rating'])) / (max(df['Rating']) - min(df['Rating']))


df = df.round(3)

df = df.sort_values('Rating', ascending=False)




ACCRankings1819 = pd.DataFrame(df['Rating'])


ACCRankings1819 = ACCRankings1819.reset_index()


ACCRankings1819.index += 1

ACCRankings1819.index.name = 'Rank'


ACCRankings1819 = ACCRankings1819.rename(columns={'index': 'Team'})


ACCRankings1819




```

```python
ACCRankings1819.to_csv('ACCRankings1819.csv')
```
